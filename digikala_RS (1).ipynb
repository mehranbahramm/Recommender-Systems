```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_rate = pd.read_csv('digikala-comments.csv')
df_user=pd.read_csv("digikala-products.csv")
```

```python
df_user_rate=pd.merge(df_rate[["id","rate","product_id"]],df_user[["id","title_fa"]],left_on="product_id",right_on="id",how="inner")
```

```python
df_user_rate = df_user_rate[["id_x", "rate", "title_fa"]]
df_user_rate.rename(columns={"id_x": "user_id"}, inplace=True)
```

```python
df_user_rate.head()
```

```python
user_rate_matrix=df_user_rate.pivot_table(index="user_id",columns="title_fa",values="rate")
user_rate_matrix.head()
```

```python
# Fill missing ratings with 0
user_rate_matrix.fillna(0,inplace=True)
user_rate_matrix.head()
```

```python
user_rate_matrix = user_rate_matrix.astype(float)
```

```python
# Compute user similarity matrix using cosine similarity (user-user matrix)
similarity_matrix=cosine_similarity(user_rate_matrix.values)
# Convert to DataFrame for easier handling
similarity_df=pd.DataFrame(similarity_matrix,index=user_rate_matrix.index,columns=user_rate_matrix.index)
similarity_df.head()
```
