```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
rating_file= '/content/ml-1m/ratings.dat'
movie_file= '/content/ml-1m/movies.dat'
```

```python
rating_columns=['UserID','MovieID','Rating','Timestamp']
rating_df=pd.read_csv(rating_file,sep='::',names=rating_columns)
rating_df
```

```python
movie_columns=["MovieID","Title","Genres"]

movie_columns_df=pd.read_csv(movie_file,sep='::',names=movie_columns, encoding='latin-1')
```

```python
rating_df.head()
```

```python
movie_columns_df.head()
```

```python
movie_rating_df=pd.merge(rating_df,movie_columns_df,on="MovieID")
movie_rating_df.drop(["Timestamp","Genres"],axis=1,inplace=True)

```

```python
movie_rating_df.head()
```

```python
movie_rating_df.isnull().sum()
```

```python
user_item_metrix= movie_rating_df.pivot_table(index="UserID",columns="Title",values="Rating")
user_item_metrix
```

```python
user_item_metrix.fillna(0,inplace=True)
```

```python
similarity_matrix=cosine_similarity(user_item_metrix)
```

```python
similarity_df= pd.DataFrame(similarity_matrix,index=user_item_metrix.index,columns=user_item_metrix.index)
similarity_df.head()
```

```python
def recommend_movie_user(UserID,n_recommendation=5,similarity_threshold=0.5):
# Get the list of similar users
 similar_users = similarity_df.loc[UserID].sort_values(ascending=False)
 similar_users=similar_users[similar_users > similarity_threshold]

# Get watched movies for the user
 user_movies= user_item_metrix.loc[UserID]
 watch_movies=user_movies[user_movies > 0].index.tolist()

# Get movies watched by similar users
 similar_users_rating=user_item_metrix.loc[similar_users.index]
 similar_users_rating=similar_users_rating.mean().sort_values(ascending=False)

# Recommend movies not yet watched by the user
 recommendations = [movie for movie in similar_users_rating.index if movie not in watch_movies]

 return recommendations[:n_recommendation]

```

```python
collaborative_recommendations_user=recommend_movie_user(UserID=11, n_recommendation=5)

collaborative_recommendations_user
```
